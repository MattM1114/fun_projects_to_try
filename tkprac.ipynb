```python
# simple calculator with  addition, subtraction, multiplication, division  
# buttons and  a display screen and  a clear button 
import tkinter as tk

# create the  main window
root = tk.Tk()
root.title("simple  calculator")
# entry widget for displaying numbers
entry = tk.Entry(root, width=16,  borderwidth=5, justify='right') 
entry.grid(row=0,column=0,columnspan=4,padx=10,pady=10)


#golbal variables
num1 = None
ops = None

# function for number button clicks 
def  button_click(num):
    current = entry.get()
    entry.delete(0, tk.END)
    entry.insert(0, str(current) + str(num))

# function for clear button
def button_click_clear():
    entry.delete(0, tk.END)

        
#  function for operator button clicks
def button_click_ops(op):
    global num1,ops
    num1 = float(entry.get())
    ops = op
    entry.delete(0, tk.END)
    
#  function for equals button click
def button_click_equals():
    global num1,ops
    num1 = float(entry.get())
    num2 = float(entry.get())
    entry.delete(0, tk.END)
    
    if ops =="+":
        entry.insert(0, str(num1 + num2))
    elif  ops == "-":
        entry.insert(0, str(num1 - num2))
    elif ops == "*":
        entry.insert(0, str(num1 * num2))
    elif ops == "/":
        if num2 != 0:
            entry.insert(0, str(num1 / num2))
        else:
            entry.insert(0,  "Error")
            
# button layout
button_text = [
    ('7',1,0),('8',1,1),('9',1,2),('/',1,3),
    ('4',2,0),('5',2,1),('6',2,2),('*',2,3),
    ('1',3,0),('2',3,1),('3',3,2),('-',3,3),
    ('0',4,0),('C',4,1),('=',4,2),('+',4,3)
]


# Create buttons and add them to the grid
for (text, row, col) in button_text:
    if text.isdigit():  # Number buttons
        button = tk.Button(root, text=text, padx=20, pady=20, font=("Arial", 15),
                           command=lambda t=text: button_click(t))
    elif text == "C":  # Clear button
        button = tk.Button(root, text=text, padx=20, pady=20, font=("Arial", 15),
                           command=button_click_clear)
    elif text == "=":  # Equals button
        button = tk.Button(root, text=text, padx=20, pady=20, font=("Arial", 15),
                           command=button_click_equals)
    else:  # Operator buttons
        button = tk.Button(root, text=text, padx=20, pady=20, font=("Arial", 15),
                           command=lambda op=text: button_click_ops(op))
    
    button.grid(row=row, column=col)

root.mainloop()
```
